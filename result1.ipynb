```python
import numpy as np
import matplotlib.pyplot as plt
import os
def str_to_dict(s):
    import re
    x=re.split(",|:" ,s[1:-2])
    res=dict()
    res[x[0][1:-1]]=float(x[1])
    res[x[2][2:-1]]=float(x[3])
    return res

def plot_ris(dir,i,axs,name,f,size=2000):
        a=open(dir,'r')
        a=a.read().split('\n')[:-1]
        a=[str_to_dict(b) for b in a]
        train=[s['train_acc'] for s in a]
        test=[s['test_acc'] for s in a]
        ge=[t1-t2 for t1,t2 in zip(train,test)]

        ge={}
        max_v=0
        for s in a:
            if s['test_acc']<max_v:
                continue
            max_v=s['test_acc']
            if s['test_acc'] in ge:
                ge[s['test_acc']]=min(ge[s['test_acc']],s['train_acc'])
            else:
                ge[s['test_acc']]=s['train_acc']
        axs[i,0].plot(np.arange(size),train[:size],label=f)
        axs[i,1].plot(np.arange(size),test[:size],label=f)
        axs[i,2].plot([v for v in ge.keys()],[k for k in ge.values()],label=f)
        axs[i,0].set_title(name+" train result")
        axs[i,0].legend()
        axs[i,1].set_title(name+" test result")
        axs[i,1].legend()
        axs[i,2].set_title(name+" test result")
        axs[i,2].legend()
        return max(test)
```

```python
dir='logs/CrossEntropyLoss/L2/Overall/OneLayerConvNet/'
size=500
files = os.listdir(dir)
fig, axs = plt.subplots(2,3,figsize=(40,15))
max_res=0
max_f=''
for f in files:
    if f=='logs':
        continue
    max_v=plot_ris(dir+f,0,axs,'',f,size=size)
    if max_v>max_res:
        max_res=max_v
        max_f=f
plot_ris(dir+max_f,1,axs,'overall',max_f,size=size)
plt.show()
```
